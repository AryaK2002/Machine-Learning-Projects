Popularity Based Recommendation

```python
rating_with_name=ratings.merge(books,on='ISBN')
```

```python
rating_with_name
```

```python
num_rating=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating
```

```python
num_rating.rename(columns={'Book-Rating':'NoOfRatings'},inplace=True)
num_rating
```

```python
avg_rating=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'AvgRating'},inplace=True)
avg_rating
```

```python
num_avg_rating=num_rating.merge(avg_rating,on='Book-Title')
num_avg_rating
```

```python
popular_rating=num_avg_rating[num_avg_rating['NoOfRatings']>=250].sort_values('AvgRating',ascending=False).head(50)
```

```python
popular_rating
```

```python
popular_rating=popular_rating.merge(books,on='Book-Title').drop_duplicates('Book-Title')
popular_rating
```

```python
popular_rating=popular_rating[['Book-Title','Book-Author','Image-URL-M','NoOfRatings','AvgRating']]
popular_rating
```

**Colaborative Filtering Recommender System**





```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
avid_readers = x[x].index
avid_readers
```

```python
filtered_rating = ratings_with_name[rating_with_name['User-ID'].isin(avid_readers)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Outlander')
```
